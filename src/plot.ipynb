```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
with open('../results/loss.pkl', 'rb') as f:
    data = pickle.load(f)

train_loss_arr = data['train_loss_arr']
valid_loss_arr = data['valid_loss_arr']       

x_grid = np.arange(len(train_loss_arr))
x = np.argmin(valid_loss_arr)
y = np.min(valid_loss_arr)
plt.plot(x_grid, train_loss_arr, label = "train loss")
plt.plot(x_grid, valid_loss_arr, label = "dev loss")
plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.ylim(0,8)
plt.vlines(x, ymin=0, ymax=y, color='k', linestyle='--')
plt.scatter(x,y, c='orangered')
plt.text(x+1.5, y-0.5,f'({x},{y:.2f})', fontsize=14)
plt.title('Transformer')
plt.show()
```
